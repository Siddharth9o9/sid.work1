# Statistical Insights


#SALES

```python
# Insights based on Cities
grouped= df.groupby("City")
#Aggregated Sales per city
agg_sales=grouped['Sales'].agg(np.sum).sort_values(ascending=False).reset_index()
#Cities with highest total sales
agg_sales.head(10)
```

```python
#Let's get some insights based on Cities
grouped= df.groupby("City")
#Aggregated Sales per city
agg_sales=grouped['Sales'].agg(np.sum).sort_values(ascending=True).reset_index()
#Cities with lowest total sales
agg_sales.head()
```

Marketing techniques could be used and logistics and supply chain should be improved after further analysis of why these Cities have less sales.

#PROFIT

```python
#Aggregated Profit per city
agg_profit=grouped['Profit'].agg(np.sum).sort_values(ascending=False).reset_index()
#Cities with Highest total Profit
agg_profit.head(10)
```

Though Philadelphia, Houston, Chicago have high sales, there profit are not in top 10. There could be a possibility of inefficient operations management and supply chain or high cost is incured. 

```python
grouped= df.groupby("City")
#Aggregated Profit per city
agg_sales=grouped['Profit'].agg(np.sum).sort_values(ascending=True).reset_index()
#Cities with lowest total profit
agg_sales.head()
```

Though the sales are significantly high in Philadelphia and Houston (in top 10), they operations in these cities are incurring losses. Operations and supply chain management should be improved and more efficiency should be attained. Inventory management is also equally important.

#DISCOUNT

```python
#Average Discount per city
avg_dist=grouped['Discount'].agg(np.mean).sort_values(ascending=False).reset_index()
#Cities with highest Average discount
avg_dist.head(10)
```

The cities which topped the Sales and Profit chart are not leading here
